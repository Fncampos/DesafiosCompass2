```python
############################################################################################
#               Funções do Processo carregamento e limpeza
#############################################################################################
#Carregamento dos dados
def loadCSV(diretorio, codificacao):
    #caminho e codificacao são strings
    arquivo = open(diretorio, encoding= codificacao)
    dados = arquivo.readlines()
    return dados

#Criada para remover quebra de linha "\n" presente no fim de cada linha
#Ex: linha antes--> Actor,Total Gross,Number of Movies,Average per Movie,#1 Movie,Gross\n,
#    depois--> Actor,Total Gross,Number of Movies,Average per Movie,#1 Movie,Gross
#OBS: (Separdor:String, lista:list)
def limpeza_tiraSeparadorIndejado(separador, lista):
    listanova =[]
    for linha in lista:
        linha.split(separador)
        linha1 = linha.rstrip('\n')
        listanova.append(linha1)
    return listanova

#Gera nova lista com um dado alterado
#OBS: (lista:list, dadoAtual:String, dadoModificado:String)
def limpeza_alteraDado(lista, dadoAtual, dadoModificado):
    listanova =[]
    for linha in lista:
        if dadoAtual in linha:
            linhanova=linha.replace(dadoAtual, dadoModificado)
            listanova.append(linhanova) 
        else:
            listanova.append(linha)
    return listanova

#Cria lista utilizando separador especificado  
#OBS: (lista:list, separador:String)
def fazlista_separdorEspecificado(lista, separador): 
    listanova =[]
    for linha in lista:
        listanova.append(linha.split(separador))
    return listanova

#Remove apenas o titulo
#OBS: (lista:list)
def removeTitulo(lista):
    listanova =[]
    for i in range(1,len(lista)):
        listanova.append(lista[i])
    return listanova

#Ordenar por ator
#OBS: (lista:list)
def ordenarPorAtor_asc(lista):
    listanova=sorted(lista)
    return listanova


#Exibição de lista
#OBS: (lista:list)
def exibe_Lista(lista):
    for dado in lista:
        print(dado)
#############################################################################################  
#                      Executando Funções de carregamento e limpeza
#############################################################################################     

#passo1
dadoBruto = loadCSV('actors.csv', 'utf8')
#passo2
transformaDado_semquebralinha = limpeza_tiraSeparadorIndejado("\n",dadoBruto)
#passo3
trasformaDado_dadoAlterado = limpeza_alteraDado(transformaDado_semquebralinha,'"Robert Downey, Jr."', 'Robert Downey Jr.')
#passo4
completoEmLista = fazlista_separdorEspecificado(trasformaDado_dadoAlterado, ",")
#passo5-dadolimpoparaAnalise
listaSemTitulo = removeTitulo(completoEmLista)
#passo6 -ordenar por ATOR
ListaPronta=ordenarPorAtor_asc(listaSemTitulo)

#############################################################################################
#       Funções de Criação de Arquivo e gravação de dados
#############################################################################################  

def cria_arquivo_txt(nomeArquivo):
    titulo=nomeArquivo
    open(titulo, "w")


def grava_lista_em_arquivo(lista, nomeArquivo):
    f = open(nomeArquivo, "a")
    for linha in lista:
        f.write(f'{str(linha)}\n')
    f.close

def grava_str_em_arquivo(texto, nomeArquivo):
    f = open(nomeArquivo, "a")
    f.write(texto)
    f.close()

#############################################################################################  
#     #1.O ator/atriz com maior número de filmes e o respectivo número de filmes.                  
############################################################################################# 

#Transforma numeros para tipos numericos
#OBS: (lista:list)
def atorComMaisFilmes(lista):
    listanova =[]
    for i in range(len(lista)):
        dadoValor = [lista[i][0],int(lista[i][2])]
        listanova.append(dadoValor)
    desc_lista_atorFilme = sorted(listanova, key=lambda filme: filme[1], reverse=True)  
    maior = desc_lista_atorFilme[0]
    return f'O Ator {maior[0]} possui maior numero de filmes, num total de {maior[1]} filmes'
    
#Chamando Função
q1 = atorComMaisFilmes(ListaPronta)

#print(q1)

#Criação e gravação de arquivo com a solução
#cria_arquivo_txt('Q1_desafio.txt')
#grava_str_em_arquivo(q1,'Q1_desafio.txt')

#############################################################################################  
#     #2.A média do número de filmes por ator.       
############################################################################################# 

#Calcula média de filmes por ator
#OBS: (lista:list)
def mediaFilmesporAtor(lista):
    listanova =[]
    soma=0
    for i in range(len(lista)):
       soma = soma + int(lista[i][2])
    media = soma/len(lista)
    return f'A media de filmes por ator e {media}'
   
q2 = mediaFilmesporAtor(ListaPronta)

#print(q2)

#Criação e gravação de arquivo com a solução
#cria_arquivo_txt('Q2_desafio.txt')
#grava_str_em_arquivo(q2,'Q2_desafio.txt')

#############################################################################################  
#     #3.O ator/atriz com a maior média por filme.       
############################################################################################# 

#Lista completa ator x media bruta
#OBS: (lista:list)
def listacompleta_atorXmediaBruta(lista):
    listanova =[]
    for i in range(len(lista)):
        ator= lista[i][0]
        avgGross= float(lista[i][3])
        listanova.append([ator, avgGross])
    desc_avgGross = sorted(listanova, key=lambda filme: filme[1], reverse=True)
    return desc_avgGross

def ator_maior_mediaBruta(lista):   
    listanova =[]
    for i in range(len(lista)):
        ator= lista[i][0]
        media= lista[i][1]
        listanova.append([ator, media])
    return f'{listanova[0][0]} e o ator com  maior media por filmes, igual a {listanova[0][1]}' 

#Chamando Função
q3_passo1 = listacompleta_atorXmediaBruta(ListaPronta)
q3_passo2 = ator_maior_mediaBruta(q3_passo1)

#print(q3_passo2)

#Criação e gravação de arquivo com a solução
#cria_arquivo_txt('Q3_desafio.txt')
#grava_str_em_arquivo(q3_passo2,'Q3_desafio.txt')
#############################################################################################  
#     #4.O nome do(s) filme(s) mais frequente(s) e sua respectiva frequência.       
############################################################################################# 

#Cria lista de filmes sem repetir filmes
#OBS: (lista:list)
def listafilmes_SemRepeticao(lista):
    listanova =[]
    soma=0
    for i in range(len(lista)):
        listanova.append(lista[i][4])
    listanova=list(set(listanova))
    listanova=sorted(listanova)
    return listanova   

#Lista completa apenas com a coluna filmes
#OBS: (lista:list)
def listafilmes_completa_colunaFilmes(lista):
    listanova =[]
    for i in range(len(lista)):
        listanova.append(lista[i][4])
    listanova=sorted(listanova)
    return listanova   

#Lista final com filmes e Frequencias
#OBS: (listaCompleta,listafilmes: list)
def filmeporFrequencia(listaCompleta,listafilmes):
    desc_freqFilme=[]
    listanova =[]
    valor=0
    
    for filme in listafilmes:
        count=0
        for filme2 in listaCompleta:
            if filme==filme2:
                count+=1
        valor=count        
        listanova.append([filme, valor]) 
        desc_freqFilme = sorted(listanova, key=lambda filme: filme[1], reverse=True)  
    return desc_freqFilme

#Chamando Funções
q4_passo1 = listafilmes_completa_colunaFilmes(ListaPronta)
q4_passo2 = listafilmes_SemRepeticao(ListaPronta)
q4_passo3 = filmeporFrequencia(q4_passo1, q4_passo2)

#exibe_Lista(passo3)

#Criação e gravação de arquivo com a solução
#cria_arquivo_txt('Q4_desafio.txt')
#grava_lista_em_arquivo(q4_passo3,'Q4_desafio.txt')
############################################################################################# 
#   #5.A lista dos Atores ordenada por pagamento. Do mais bem pago para o menos bem pago     
############################################################################################# 

#Lista completa atores mais bem pagos está na questão 3
#Chamando Função
#Na linha 158: q3_passo1 = listacompleta_atorXmediaBruta(ListaPronta)

exibe_Lista(q3_passo1)

#Criação e gravação de arquivo com a solução
#cria_arquivo_txt("Q5_desafio.txt")
#grava_lista_em_arquivo(q3_passo1, 'Q5_desafio.txt')
#############################################################################################
```
